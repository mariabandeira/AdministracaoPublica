### Recuperando os dados

```python
import numpy as np
import pandas as pd
```

```python
FILE_PATH_2018 = './BasesRN/cadunico_rn_2018.csv'
cadUnicoRN2018 = pd.read_csv(FILE_PATH_2018, sep=';', low_memory=False)
```

```python
FILE_PATH_2017 = './BasesRN/cadunico_rn_2017.csv'
cadUnicoRN2017 = pd.read_csv(FILE_PATH_2017, sep=';', low_memory=False)
```

```python
FILE_PATH_2016 = './BasesRN/cadunico_rn_2016.csv'
cadUnicoRN2016 = pd.read_csv(FILE_PATH_2016, sep=';', low_memory=False)
```

```python
# Juntando os dois datasets
cadUnicoRN = pd.concat([cadUnicoRN2018, cadUnicoRN2017, cadUnicoRN2016], ignore_index=True)
```

### Pré-processamento dos dados

```python
# removendo colunas que não se encontram em um formato adequado para o treinamento de ML
cadUnicoRN = cadUnicoRN.drop(columns=['dat_cadastramento_fam', 'dat_alteracao_fam', 'dat_atualizacao_familia', 'nom_estab_assist_saude_fam', 'cod_eas_fam', 'nom_centro_assist_fam', 'cod_centro_assist_fam'])

cadUnicoRN.head()
```

```python
# verificando linhas duplicadas devido a mesclagem de diferentes datasets
duplicados = cadUnicoRN.duplicated(subset=['id_familia'], keep=False)

# print("Registros dublicados encontrados");
# print(cadUnicoRN[duplicados])

cadUnicoRN = cadUnicoRN.drop_duplicates(subset=['id_familia'], keep='last')
```

```python
# colunas que precisam de tratamento de tipos
columns_to_convert = ['vlr_renda_media_fam', 'peso.fam']

for column in columns_to_convert:
    cadUnicoRN[column] = cadUnicoRN[column].replace(',', '.', regex=True).astype(float)

# verificando valores ausentes em cada coluna
missing_values = cadUnicoRN.isnull().sum()

print('As colunas com valores ausentes são:\n', (missing_values))
```

```python
columns_to_remove = ['cod_escoa_sanitario_domic_fam', 'ind_parc_mds_fam', 'id_familia']

cadUnicoRN = cadUnicoRN.drop(columns=columns_to_remove, axis=1)

# removendo linhas com valores ausentes
cadUnicoRN = cadUnicoRN.dropna()

missing_values = cadUnicoRN.isnull().sum()
print('As colunas com valores ausentes são:\n', (missing_values))
```

```python
feature_names = cadUnicoRN.columns.tolist()
feature_names.remove('marc_pbf')
```

```python
# identificando colunas com apenas um valor
single_value_columns = [col for col in cadUnicoRN.columns if cadUnicoRN[col].nunique() == 1 | 0]

print(f"Colunas com um único valor: {single_value_columns}")

cadUnicoRN = cadUnicoRN.drop(columns=single_value_columns)
```

```python
cadUnicoRN.info()
```

```python
# removendo colunas que ainda não se encontram em um formato adequado para o treinamento de ML
cadUnicoRN = cadUnicoRN.drop(columns=['cd_ibge', 'estrato', 'peso.fam'])
```

```python
def classificar(x):
    if x == 1:
        return 1
    elif x == 2:
        return 2
    elif x == 3:
        return 3
    elif x == 4:
        return 4
    elif x == 5:
        return 5
    else:
        return 6

cadUnicoRN['qtde_pessoas'] = cadUnicoRN['qtde_pessoas'].apply(classificar)
cadUnicoRN['qtde_pessoas'].value_counts()
```
