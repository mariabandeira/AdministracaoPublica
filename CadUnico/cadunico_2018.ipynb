```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cadunico_df = pd.read_csv('/home/aluno/cadunico/base_amostra_familia_201812.csv', sep=';')
```

```python
cadunico_df.head()
```

```python
cadunico_df.describe()
```

```python
cadunico_df.info()
```

```python
# codigo do ibge de joão pessoa: 2507507
caduni_jp = cadunico_df[cadunico_df['cd_ibge'] == 2507507]

caduni_jp.head()
```

```python
caduni_jp.reset_index(inplace=True)

caduni_jp
```

```python
caduni_jp.describe()
```

```python
caduni_jp.info()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=caduni_jp, x='qtde_pessoas', y='vlr_renda_media_fam', s=100)

# Adicionando título e rótulos
plt.title('Relação entre Renda Familiar e Quantidade de Pessoas na Família')
plt.xlabel('Quantidade de Pessoas na Família')
plt.ylabel('Renda Familiar (R$)')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```

```python
# marc_pbf : se a familia é beneficiaria do bolsa familia ou não

plt.figure(figsize=(8, 5))
sns.countplot(data=caduni_jp, x='marc_pbf', hue='marc_pbf', palette='viridis')

# Adicionando título e rótulos
plt.title('Famílias beneficiárias do Bolsa Família')
plt.grid(axis='y')

# Exibindo o gráfico
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=caduni_jp, x='marc_pbf', y='vlr_renda_media_fam', s=100)

# Adicionando título e rótulos
plt.title('Relação entre Renda Familiar e Benefício do Bolsa Família')
plt.xlabel('Benefício Bolsa Família')
plt.ylabel('Renda Familiar (R$)')
plt.grid(True)

# Exibindo o gráfico
plt.show()
```
