## Mesclando os dados

```python
# Adiciona a coluna ano em cada base
for ano, df in [
    (2012, cadunico_2012),
    (2013, cadunico_2013),
    (2014, cadunico_2014),
    (2015, cadunico_2015),
    (2016, cadunico_2016),
    (2017, cadunico_2017),
    (2018, cadunico_2018),
]:
    df["ano"] = ano
```

```python
# Mesclando os arquivos CSV em um único DataFrame
cadunico_combined = pd.concat([cadunico_2012, cadunico_2013, cadunico_2014,
                              cadunico_2015, cadunico_2016, cadunico_2017, cadunico_2018],
                              ignore_index=True)
```

```python
# Filtrando os dados apenas para os municípios do RN
cadunico_ce_combined = cadunico_combined[cadunico_combined['cd_ibge'].astype(str).str.startswith('23')]

# Salvando o arquivo filtrado
cadunico_ce_combined.to_csv('cadunico_ce_combined.csv', index=False, sep=';')
```
