# Data Cleaning for Piauí

This notebook is designed to clean and preprocess the Cadastro Único data for the state of Piauí (PI). The goal is to ensure that the data is in a suitable format for analysis and modeling.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Define the path to the raw data
data_path = 'BaesesPI/cadunico_pi_clean.csv'

# Load the dataset
data = pd.read_csv(data_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Fill or drop missing values as necessary
# Example: Filling missing values with the mean for numerical columns
for column in data.select_dtypes(include=[np.number]).columns:
    data[column].fillna(data[column].mean(), inplace=True)

# Example: Dropping rows with missing categorical values
data.dropna(subset=['categorical_column'], inplace=True)
```

```python
# Convert data types if necessary
data['date_column'] = pd.to_datetime(data['date_column'])

# Display the cleaned data
data.info()
```

```python
# Save the cleaned dataset
cleaned_data_path = 'BaesesPI/cadunico_pi_cleaned.csv'
data.to_csv(cleaned_data_path, index=False)

# Confirm that the cleaned data has been saved
print(f'Cleaned data saved to {cleaned_data_path}')
```
