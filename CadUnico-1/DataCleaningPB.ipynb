# Data Cleaning for Piauí

This notebook is designed to clean and preprocess the Cadastro Único data for the state of Piauí (PI). The goal is to prepare the data for further analysis and modeling.

```python
import pandas as pd
import numpy as np

# Load the raw data
data_path = 'BaesesPI/cadunico_pi_clean.csv'
data = pd.read_csv(data_path)

# Display the first few rows of the dataset
data.head()
```

```python
# Check for missing values
missing_values = data.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Fill missing values or drop rows/columns as necessary
# Example: Filling missing values with the mean for numerical columns
for column in data.select_dtypes(include=[np.number]).columns:
    data[column].fillna(data[column].mean(), inplace=True)

# Example: Dropping rows with missing categorical values
data.dropna(subset=['categorical_column'], inplace=True)
```

```python
# Convert categorical variables to category type
categorical_columns = ['categorical_column1', 'categorical_column2']
for column in categorical_columns:
    data[column] = data[column].astype('category')
```

```python
# Save the cleaned data
cleaned_data_path = 'BaesesPI/cadunico_pi_cleaned.csv'
data.to_csv(cleaned_data_path, index=False)
print('Cleaned data saved to:', cleaned_data_path)
```
