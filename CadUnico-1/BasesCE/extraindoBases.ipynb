# Extraindo Bases - Piauí

Este notebook é utilizado para extrair e processar dados do Cadastro Único (CadÚnico) para o estado do Piauí.

```python
import pandas as pd
import os

# Definindo o caminho para os dados
data_path = 'BasesPI/BasesAno/'

# Listando os arquivos disponíveis
files = os.listdir(data_path)
files
```

```python
# Função para carregar e processar os dados
def load_and_process_data(year):
    file_name = f'cadunico_pi_{year}.csv'
    file_path = os.path.join(data_path, file_name)
    df = pd.read_csv(file_path)
    # Aqui você pode adicionar o processamento necessário
    return df

# Carregando dados de exemplo para o ano de 2018
data_2018 = load_and_process_data(2018)
data_2018.head()
```

```python
# Exemplo de análise: contagem de famílias por município
familias_por_municipio = data_2018['cd_ibge'].value_counts()
familias_por_municipio
```

```python
# Salvar os dados processados em um novo arquivo CSV
output_path = 'cadunico_pi_clean.csv'
data_2018.to_csv(output_path, index=False)
print(f'Dados processados e salvos em: {output_path}')
```
