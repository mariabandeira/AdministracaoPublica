# Análise do Cadastro Único - 2018

Este notebook contém a análise dos dados do Cadastro Único (CadÚnico) para o ano de 2018. O objetivo é explorar e visualizar as informações disponíveis, bem como realizar algumas análises estatísticas relevantes.

```python
# Importando bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurando o estilo do seaborn
sns.set(style='whitegrid')

# Carregando os dados
data_path = 'BasesCompletas/base_amostra_familia_201812.csv'
cadunico_2018 = pd.read_csv(data_path, sep=';', low_memory=False)

# Exibindo as primeiras linhas do dataset
cadunico_2018.head()
```

```python
# Análise descritiva
descricao = cadunico_2018.describe(include='all')
descricao
```

```python
# Visualizando a distribuição da renda média
plt.figure(figsize=(10, 6))
sns.histplot(cadunico_2018['vlr_renda_media_fam'], bins=30, kde=True)
plt.title('Distribuição da Renda Média das Famílias - 2018')
plt.xlabel('Renda Média (R$)')
plt.ylabel('Frequência')
plt.show()
```

```python
# Analisando a relação entre renda média e número de pessoas na família
plt.figure(figsize=(10, 6))
sns.scatterplot(x='qtde_pessoas', y='vlr_renda_media_fam', data=cadunico_2018)
plt.title('Relação entre Renda Média e Número de Pessoas na Família - 2018')
plt.xlabel('Número de Pessoas')
plt.ylabel('Renda Média (R$)')
plt.show()
```

```python
# Salvando o dataframe limpo para uso futuro
cadunico_2018.to_csv('cadunico_2018_clean.csv', index=False, sep=';')
```
